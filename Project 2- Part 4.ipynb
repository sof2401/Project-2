```python
search = tmdb.Discover()

first_page_movies = search.movie(page=1)['results']

titles = []
imdb_ids = []
budgets = []
revenues = []
certifications = []

for movie_info in first_page_movies:
    movie_details = tmdb.Movies(movie_info['id']).info()
    
    titles.append(movie_info['title'])
    imdb_ids.append(movie_details.get('imdb_id'))
    budgets.append(movie_details.get('budget'))
    revenues.append(movie_details.get('revenue'))
    
    certification = (
        movie_details.get('release_dates', {}).get('results', [{}])[0]
        .get('release_dates', [{}])[0]
        .get('certification')
    )
    certifications.append(certification)

tmdb_data= pd.DataFrame({
    'imdb_id': imdb_ids,
    'budget': budgets,
    'revenue': revenues,
    'certification': certifications
})

print(tmdb_data)
```
