```python
import pandas as pd
import numpy as np
import os
os.makedirs('Data/', exist_ok=True)
```

```python
os.listdir("Data/")
```

```python
akas = pd.read_csv("Data/title-akas-us-only.csv", sep=',', low_memory=False)
```

```python
basics = pd.read_csv("Data/title.basics.tsv.gz", sep='\t', low_memory=False)
basics
```

```python
print (basics.columns)
```

```python
filter_us_titles = basics['tconst'].isin(akas['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
basics= basics.replace({'\\N': np.nan})
```

```python
basics['runtimeMinutes']= basics['runtimeMinutes'].dropna(inplace=True)
```

```python
basics['genres']= basics['genres'].dropna(inplace=True)
```

```python
basics = basics[basics['titleType'] == 'movie']
basics
```

```python
basics['startYear']= basics['startYear'].astype(float)
```

```python
basics = basics[(basics['startYear'] >= 2000) & (basics['startYear'] <= 2022)]
```

```python
basics
```

```python
filter_ratings = basics['genres'].str.contains('Documentary')
```
